**Using Breast Cancer Wisconsin dataset**

```python
# Import libraries
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

- Import all the necessary Python libraries for loading data, splitting data, building models, evaluating them, and plotting results.

- seaborn is imported but not used in this snippet (often used for advanced visualization).

```python
# Load the Breast Cancer dataset
data = load_breast_cancer()
X = pd.DataFrame(data.data, columns=data.feature_names)
y = data.target  # 0 = malignant, 1 = benign
```

- Load the breast cancer dataset from scikit-learn.

- Create a pandas DataFrame X with the features and their column names for easier manipulation and visualization.

- Extract the target variable y, which contains labels: 0 means malignant tumor, 1 means benign.

```python
# Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

- Split the dataset into training (70%) and testing (30%) subsets.

- random_state=42 ensures reproducibility of the split.

```python
# Train Decision Tree
tree = DecisionTreeClassifier(max_depth=4, random_state=0)
tree.fit(X_train, y_train)
y_pred_tree = tree.predict(X_test)
```

- Initialize a Decision Tree classifier with max depth of 4 to control model complexity.

- Train (fit) the model on the training data.

- Predict the labels for the test data.

```python
# Train Random Forest
forest = RandomForestClassifier(n_estimators=100, random_state=0)
forest.fit(X_train, y_train)
y_pred_forest = forest.predict(X_test)
```

- Initialize a Random Forest classifier with 100 trees.

- Train the Random Forest on the training data.

- Predict labels for the test data using the trained Random Forest.

```python
# Evaluate both models
acc_tree = accuracy_score(y_test, y_pred_tree)
acc_forest = accuracy_score(y_test, y_pred_forest)
```

```python
print("Decision Tree Accuracy:", acc_tree)
print("Random Forest Accuracy:", acc_forest)
```

- Calculate accuracy for both models by comparing predicted labels with true test labels.

- Print the accuracy scores for Decision Tree and Random Forest.



```python
# Visualize Decision Tree
plt.figure(figsize=(20, 10))
plot_tree(tree, feature_names=data.feature_names, class_names=data.target_names, filled=True)
plt.title("Decision Tree for Breast Cancer Diagnosis")
plt.show()
```

- Create a large figure for plotting.

- Visualize the trained Decision Tree with feature names and class names displayed.

- Use color (filled=True) to make the plot more readable.

- Add a title and display the plot.

```python
# Classification report (optional for deeper analysis)
print("Decision Tree Classification Report:")
print(classification_report(y_test, y_pred_tree))
```

```python
print("Random Forest Classification Report:")
print(classification_report(y_test, y_pred_forest))
```

- Print detailed classification reports for both models, including precision, recall, F1-score, and support for each class.

- Helps understand model performance beyond accuracy, especially for imbalanced classes.



**Using Iris Dataset**

```python
# Import libraries
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
import matplotlib.pyplot as plt
import pandas as pd

# Load the Iris dataset
data = load_iris()
X = pd.DataFrame(data.data, columns=data.feature_names)
y = data.target  # 0, 1, 2 (three classes of iris flowers)

# Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train Decision Tree
tree = DecisionTreeClassifier(max_depth=4, random_state=0)
tree.fit(X_train, y_train)
y_pred_tree = tree.predict(X_test)

# Train Random Forest
forest = RandomForestClassifier(n_estimators=100, random_state=0)
forest.fit(X_train, y_train)
y_pred_forest = forest.predict(X_test)

# Evaluate both models
acc_tree = accuracy_score(y_test, y_pred_tree)
acc_forest = accuracy_score(y_test, y_pred_forest)
print("Decision Tree Accuracy:", acc_tree)
print("Random Forest Accuracy:", acc_forest)

# Visualize Decision Tree
plt.figure(figsize=(20, 10))
plot_tree(tree, feature_names=data.feature_names, class_names=data.target_names, filled=True)
plt.title("Decision Tree for Iris Classification")
plt.show()

# Classification report (optional for deeper analysis)
print("Decision Tree Classification Report:")
print(classification_report(y_test, y_pred_tree, target_names=data.target_names))
print("Random Forest Classification Report:")
print(classification_report(y_test, y_pred_forest, target_names=data.target_names))
```



* The code imports libraries for data loading, model building, evaluation, and visualization.
* It loads the Iris dataset into a DataFrame `X` (features) and a target array `y` (class labels for 3 iris species).
* The dataset is split into training (70%) and testing (30%) sets with a fixed random state for reproducibility.
* A Decision Tree classifier with `max_depth=4` is trained on the training data to balance complexity and avoid overfitting.
* Predictions are made on the test data using the trained Decision Tree.
* A Random Forest classifier with 100 trees is trained on the same training data.
* Predictions are made on the test data using the Random Forest.
* Accuracy scores for both classifiers are computed and printed to compare their performance.
* The Decision Tree is visualized with feature names and class names; nodes are color-filled to show class distributions.
* Classification reports for both models are printed, showing precision, recall, F1-score, and support for each Iris class.
* These reports provide detailed insights into model performance beyond simple accuracy.



**Using Digits dataset**

```python
# Import libraries
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
import matplotlib.pyplot as plt
import pandas as pd

# Load the Digits dataset
data = load_digits()
X = pd.DataFrame(data.data, columns=[f'pixel_{i}' for i in range(data.data.shape[1])])
y = data.target  # digits 0-9

# Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train Decision Tree
tree = DecisionTreeClassifier(max_depth=10, random_state=0)
tree.fit(X_train, y_train)
y_pred_tree = tree.predict(X_test)

# Train Random Forest
forest = RandomForestClassifier(n_estimators=100, random_state=0)
forest.fit(X_train, y_train)
y_pred_forest = forest.predict(X_test)

# Evaluate both models
acc_tree = accuracy_score(y_test, y_pred_tree)
acc_forest = accuracy_score(y_test, y_pred_forest)
print("Decision Tree Accuracy:", acc_tree)
print("Random Forest Accuracy:", acc_forest)

# Visualize Decision Tree (optional)
plt.figure(figsize=(20, 10))
plot_tree(tree, max_depth=2, feature_names=X.columns, class_names=[str(i) for i in data.target_names], filled=True)
plt.title("Decision Tree for Digits Classification (partial tree)")
plt.show()

# Classification report (optional for deeper analysis)
print("Decision Tree Classification Report:")
print(classification_report(y_test, y_pred_tree, target_names=[str(i) for i in data.target_names]))
print("Random Forest Classification Report:")
print(classification_report(y_test, y_pred_forest, target_names=[str(i) for i in data.target_names]))
```



* The code imports libraries for dataset loading, model training, evaluation, and visualization.
* It loads the Digits dataset, which contains 8x8 pixel images of handwritten digits (0 to 9).
* The pixel data is converted into a pandas DataFrame `X` with columns named `pixel_0`, `pixel_1`, ..., representing each pixel.
* The target vector `y` contains digit labels (0 through 9).
* The dataset is split into training (70%) and testing (30%) sets with a fixed `random_state` for reproducibility.
* A Decision Tree classifier with `max_depth=10` is trained on the training data to capture complex patterns while controlling overfitting.
* Predictions are generated on the test set using the Decision Tree.
* A Random Forest classifier with 100 trees is trained on the same training data.
* Predictions are generated on the test set using the Random Forest.
* Accuracy scores for both models are calculated by comparing predicted and true labels, and the scores are printed to compare their performance.
* The Decision Tree is visualized (showing only the top 2 levels for clarity) with pixel feature names and digit class names; nodes are color-filled for better interpretability.
* Detailed classification reports for both models are printed, showing precision, recall, F1-score, and support for each digit class (0-9).
* These reports provide deeper insights into the strengths and weaknesses of each model beyond overall accuracy.
